```python
import numpy as np

#Dicionário com cada posição do tabuleiro corresponde no array
dicio = {
    '1,1': 0, '1,2': 1, '1,3': 2, '1,4': 3, '1,5': 4,
    '2,1': 5, '2,2': 6, '2,3': 7, '2,4': 8, '2,5': 9,
    '3,1': 10, '3,2': 11, '3,3': 12, '3,4': 13, '3,5': 14,
    '4,1': 15, '4,2': 16, '4,3': 17, '4,4': 18, '4,5': 19,
    '5,1': 20, '5,2': 21, '5,3': 22, '5,4': 23, '5,5': 24
}

#Dicionário com cada movimento possível para cada peça que podem ser retiradas
dicioMovValidos = {
    '1,1': [4,20], '1,2': [0,4,21], '1,3': [0,4,22], '1,4': [0,4,24], '1,5': [0,24],
    '2,1': [0,9,20], '2,5': [4,5,24], '3,1': [0,14,20], '3,5': [4,10,24], '4,1': [0,19,20], '4,5': [4,15,24],
    '5,1': [0,24], '5,2': [1,20,24], '5,3': [2,20,24], '5,4': [3,20,24], '5,5': [4,20]
}


#Dicionário guardando as linhas, colunas e diagonais e seus respectivos indices no array tabuleiro
dicioLinhas = {
    'Linha 1': [0,1,2,3,4], 'Linha 2': [5,6,7,8,9], 'Linha 3': [10,11,12,13,14], 'Linha 4': [15,16,17,18,19], 'Linha 5': [20,21,22,23,24],
    'Coluna 1': [0,5,10,15,20], 'Coluna 2': [1,6,11,16,21], 'Coluna 3': [2,7,12,17,22], 'Coluna 4': [3,8,13,18,23], 'Coluna 5': [4,9,14,19,24],
    'Diagonal 1': [0,6,12,18,24], 'Diagonal 2': [4,8,12,16,20]
}



#Array de peças das bordas
arrPecasBordas = np.array(['1,1','1,2','1,3','1,4','1,5','2,1','2,5','3,1','3,5','4,1','4,5','5,1','5,2','5,3','5,4','5,5'])


#Função para imprimir
def imprimirTabuleiro(arr):
      return print("| " + arr[0] + " | " + arr[1] + " | " + arr[2]+ " | " + arr[3] + " | " + arr[4] + " | " + "\n" +
      "| " + arr[5] + " | " + arr[6] + " | " + arr[7]+ " | " + arr[8] + " | " + arr[9] + " | " + "\n" +
      "| " + arr[10] + " | " + arr[11] + " | " + arr[12]+ " | " + arr[13] + " | " + arr[14] + " | " + "\n" +
      "| " + arr[15] + " | " + arr[16] + " | " + arr[17]+ " | " + arr[18] + " | " + arr[19] + " | " + "\n" +
      "| " + arr[20] + " | " + arr[21] + " | " + arr[22]+ " | " + arr[23] + " | " + arr[24] + " | " + "\n")





#Função para pegar peça
def pegarPeca(arr, posPeca): #Recebe o arrya e o local de onde o jogador quer retirar a peça (LinhaXColuna)
  array = np.array(arr) #Cópia do array original
  if posPeca in arrPecasBordas: #Verifica se a peça é uma peça da borda
      indice = dicio[posPeca] #Ve no dicionario qual o indice no array corresponde a linha e a coluna retirada
      array[indice] = "❗" #Coloca o símbolo na posição que foi retirada
      imprimirTabuleiro(array) #Imprime tabuleiro com a peça faltando
  else: #Caso não seja uma peça válida que possa ser retirada
    print("Peça inválida!") #Exibe mensagem
    pegarPeca(arr) #Chama a função de novo


def moverElemento(arr, origem, destino, peca):
    arr = arr.copy()  # Evita modificar o array original

    simbolo = peca  # Pega o valor a ser movido
    arr = np.delete(arr, origem)  # Remove o valor da posição original
    arr = np.insert(arr, destino, simbolo)  # Insere na nova posição

    return arr


def inserirPeca(arr,posTirada,peca): #Insere peça no local desejado

  posPeca = input("Digite a posição onde deseja colocar a peça (LinhaXColuna): ") #Recebe o local onde o jogador quer realocar a peça (LinhaXColuna)
  arrMovValidos = dicioMovValidos[posTirada] #Cria um array com os valores do dicionário de movimentos válidos para aquela peça
  posPeca = dicio[posPeca] #Recebe do dicionário o valor do indice do array corresponde a posição que o jogador deseja inserir
  posTirada = dicio[posTirada] #Recebe do dicionário o valor do indice do array corresponde a posição onde a peça foi retirada
  if posPeca in arrMovValidos: #Verifica se a peça que foi retirada pode ser colocada na posição que o jogador deseja
    arr = moverElemento(arr,posTirada,posPeca, peca) #cria um array novo com a peça no lugar desejado e as outras peças movidas para suas novas posições
    return arr #Retorna o novo array tabuleiro
  else: #Verifica que a peça não pode ser inserida no local que o jogador deseja
    print("Movimento inválido")
    inserirPeca(arr,posTirada) #Chama novamente a função para que o jogador faça uma nova escolha válida


pontos = 0

def testarObjetivo(arrTabuleiro, dicioLinhas): #Recebe tabuleiro
  for linha, indices in dicioLinhas.items(): #Percorre linhas e indices no dicionario de linhas e colunas
    valores = [arrTabuleiro[i] for i in linha]  # Obtém os valores correspondentes do array de cada linha, coluna ou horizontal usando os índices
    if all(v == "❌" for v in valores):  # Verifica se todos os elementos na linha são "❌"
      print("O jogador 2 ganhou!") #Se sim, avisa que jogador 2 ganhou e retorna verdadeiro
      return True
    elif all(v == "🔴" for v in valores):  #Verifica se todos os elementos são "🔴"
      print("O jogador 1 ganhou!") #Se sim, avisa que jogador 1 ganhou e retorna verdadeiro
      return True
  return False #Se nenhum dos dois ganhou ainda retorna false



#Array tabuleiro
arrTabuleiro = np.full(25, "⬜", dtype=str)

print("\n\n❌🔴❌🔴❌🔴❌🔴 QUIXO ❌🔴❌🔴❌🔴❌🔴\n\n")


turno = 0
flag = False
start = input("Digite 'S' para começar: ")

if start == "S":
  flag = True
  print("\nComeçou!\n")
  imprimirTabuleiro(arrTabuleiro)
  while flag:
      if turno % 2 == 0:
        print("Vez jogador 1 (O)\n")
        peca = "🔴"
        # imprimirTabuleiro(arrTabuleiro)
        posPeca = input("Digite a peça que deseja mover (LinhaXColuna): ")
        if posPeca == "O":
          flag = False
        else:
          pegarPeca(arrTabuleiro,posPeca)
          arrTabuleiro = inserirPeca(arrTabuleiro,posPeca,peca)
          imprimirTabuleiro(arrTabuleiro)
          turno += 1
      else:
        print("Vez jogador 2 (X)\n")
        peca = "❌"

        # imprimirTabuleiro(arrTabuleiro)
        posPeca = input("Digite a peça que deseja mover (LinhaXColuna): ")
        if posPeca == "O":
          flag = False
        else:
          pegarPeca(arrTabuleiro,posPeca)
          arrTabuleiro = inserirPeca(arrTabuleiro,posPeca,peca)
          imprimirTabuleiro(arrTabuleiro)
          turno += 1
  else:
    print("Saiu do jogo")
else:
  print("Não começou")


```
